```python
import pandas as pd
import numpy as np
import os

os.makedirs('../data', exist_ok=True)
np.random.seed(42)
n_samples = 2500

thrift = np.random.normal(50, 25, n_samples)
gorpcore = np.random.normal(30, 20, n_samples)
y2k = np.random.normal(40, 25, n_samples)
avant_garde = np.random.normal(20, 15, n_samples)

gorpcore = gorpcore - (thrift * 0.2) + (avant_garde * 0.4)
y2k = y2k + (thrift * 0.3)

data = np.column_stack([thrift, gorpcore, y2k, avant_garde])
data = np.clip(data, 0, 100)

df = pd.DataFrame(data, columns=['thrift_energy', 'gorpcore_utility', 'y2k_nostalgia', 'avant_garde'])
df.to_csv('../data/drip_stats.csv', index=False)
```
